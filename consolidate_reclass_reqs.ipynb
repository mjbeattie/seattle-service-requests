This script consolidates the reclassified service request data files into one csv for uploading to the PostgreSQL database.

```python
import pandas as pd, os

path = 'C:\\Users\\mjbea\\OneDrive\\GitHub\\seattle-service-requests-data\\Reclassified Requests\\'

# Initialize an empty list of dataframes
dfs = []
for i in range(0,62):
    filename = path + 'reclassified_fifi_reqs_' + str(i) + '.csv'
    df = pd.read_csv(filename, index_col=None, encoding="utf-8", header=0)
    dfs.append(df)
    
# Concatenate all the dataframes into one
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
big_frame
```

```python
# Save the dataframe to a new csv
outfilename = path + 'all_reclass_fifi_reqs.csv'
big_frame.to_csv(outfilename, sep="|", index=False, encoding="utf-8")
```
